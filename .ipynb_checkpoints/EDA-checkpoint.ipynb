```python
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
```

```python
data = pd.read_excel("pct_data/PCT Water_ Campo Archive 2019-09-09 07_13_13.xlsx",usecols=[2,3,4,5]).dropna()
data.columns = ["loc_id", "loc", "description", "date"]
data = data.apply(lambda x: x.astype(str).str.lower())
data.tail(20)
```

```python

data["loc"] = data["loc"].str.replace(r".+?(?=\\)","")
print(data.tail(20))
data["loc"] = data["loc"].str.replace(r"\[.*\]","")
data["loc"] = data["loc"].str.replace(r"\(.*\)","")
data["loc"] = data["loc"].str.replace(r"\{.*\}","")

data["loc"] = data["loc"].str.replace('[^a-zA-Z]', '')
```

```python
flowing_keywords = ["good", "flowing", "lots"]
def isFlowing(descriptions):
    return [any([kw in description for kw in flowing_keywords]) for description in descriptions]

not_flowing_keywords = ["bad", "dry", "no water", ]
def isFlowing(descriptions):
    return [any([kw in description for kw in flowing_keywords]) for description in descriptions]
```

```python
data["flowing"] = isFlowing(data["description"])
data["not_flowing"] = isNotFlowing(data["description"])
```
